```python
import turtle
import time
scr=turtle.getscreen()
scr.bgcolor("black")
bug=turtle.Turtle()
scr.title("Python Turtle Graphics")
bug.penup()
bug.speed(0)

def draw_rectangle(length,height,color):
    bug.pendown()
    bug.color(color)
    bug.begin_fill()
    bug.fd(length)
    bug.lt(90)
    bug.fd(height)
    bug.lt(90)
    bug.fd(length)
    bug.lt(90)
    bug.fd(height)
    bug.end_fill()
def draw_stripe(y,color):
    bug.penup()
    bug.goto(-1.9*height/2,y)
    bug.lt(90)
    bug.pendown()
    draw_rectangle(length,height/13, color)
def main():
    def draw_flag(height):
        bug.goto(-1.9*height/2,0)
        draw_rectangle(1.9*height,height,"black")
        def draw_stripe(y,color):
            bug.penup()
            bug.goto(-1.9*height/2,y)
            bug.lt(90)
            bug.pendown()
            draw_rectangle(1.9*height,height/13, color)
        draw_stripe(0, "red")
        draw_stripe(height/13, "white")
        draw_stripe(2*height/13, "red")
        draw_stripe(3*height/13, "white")
        draw_stripe(4*height/13, "red")
        draw_stripe(5*height/13, "white")
        draw_stripe(6*height/13, "red")
        draw_stripe(7*height/13, "white")
        draw_stripe(8*height/13, "red")
        draw_stripe(9*height/13, "white")
        draw_stripe(10*height/13, "red")
        draw_stripe(11*height/13, "white")
        draw_stripe(12*height/13, "red")
        bug.penup()
        bug.goto(-1.9*height/2,height)
        draw_rectangle(7*height/13,2*1.9*height/5,"blue")
        def draw_star(x,y,size, color):
            bug.goto(x,y)
            i=6
            bug.begin_fill()
            bug.color(color)
            while i>0:
                bug.fd(size)
                bug.rt(145)
                i=i-1
            bug.end_fill()
            bug.penup()
    
        def stars6():
            dist_of_stars=3/25*height
            dist_bet_lines=height/13 + dist_of_stars/5
            y=12*height/13
            for row in range (0,5):
                x=-1.9*height/2 + 15.5
                for stars in range(0,6):
                    draw_star(x,y,1.9/30*height, "white")
                    x=x+dist_of_stars
                y=y-dist_bet_lines
        def stars5():
            dist_of_stars=3/25*height
            dist_bet_lines= height/13 + dist_of_stars/5
            y=12*height/13 - 5
            for row in range(0,4):
                x= -1.9*height/2 + 31.5
                for star in range (0,5):
                    draw_star(x,y,1.9/30*height, "white")
                    x=x+dist_of_stars
                y=y-dist_bet_lines
        stars6()
        stars5()
        bug.hideturtle()
    draw_flag(250)

if __name__=="__main__":
    main()
```
